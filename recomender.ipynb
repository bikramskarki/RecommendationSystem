```python
#Implementation of item-item colaborative filtering for netlfix example data
#Bikram Karki

import pandas as pd
import sklearn
import numpy as np
import scipy
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Read given example dataset by our professor
dataset = pd.read_csv('exampledataset/data.txt', delimiter=',')

#Read netflix ratings given   
columns = ['IDusr','IDmve','Ratings'] 
dataset = pd.read_csv('netflix/ratings.txt', delimiter=',', names=columns)
dataset.head()
```

```python
# matrix for user-movie 
matrixUsrMovie = dataset.pivot(index ='IDusr', columns ='IDmve', values ='Ratings')
matrixUsrMovie.fillna(0, inplace=True)  # remove NaNs
matrixUsrMovie.head() 
```

```python
# cosine similarity getting using sklearn
cosine_sim_matrix = cosine_similarity(matrixUsrMovie.T, matrixUsrMovie.T)
cosine_sim_matrix
```

```python
# final predict rating
def final_rating_predict(usrID, movieID):   
    
    df = pd.DataFrame(matrixUsrMovie)
    
    #Using numpy getting index of userID and movieId from the matrixUsrMovie panda dataframe
    user_pos = np.where(df.index.values==usrID)[0][0]
    movie_pos = np.where(df.columns.values==movieID)[0][0]
    
    # Rating calculation
    evaluate_rating = np.sum( np.dot(matrixUsrMovie.iloc[user_pos, :] , cosine_sim_matrix[:, movie_pos])) / (np.sum(cosine_sim_matrix[:, movie_pos]) - 1)
    finalrating = "{:.1f}".format(evaluate_rating)
    
    print ("Final Predicted rating :  ", finalrating)
```

```python
#Please enter User Id and Movie IDin the input parameters of final_rating_predict
final_rating_predict(305,1333) 
```
